```python
import numpy as np
import plotly.graph_objects as go
```

```python
def create_meshgrid(*xi):   
    meshgrid_x = np.meshgrid(*xi)
    y = np.sin(np.sqrt(np.sum(x_**2 for x_ in meshgrid_x))) + np.random.normal(loc=0, scale=0.15, size=N)

    return meshgrid_x, y
```

```python
def knn(*xi, x_new, K): 
    
    meshgrid_x, y = create_meshgrid(*xi)

    meshgrid_x_new = np.meshgrid(*x_new)

    # Calculate the Manhtattan Distance
    D = np.sum(np.abs( np.array(meshgrid_x) - np.array(meshgrid_x_new) ), axis=0)

    # Flatten the distance grid and get the indexes of the K closest points
    closest_indexes = np.argsort(D.flatten())[:K]

    # Collect the K values for xi,...,xn and y values
    xi_closest = [xi.flatten()[closest_indexes] for xi in meshgrid_x]
    y_closest = y.flatten()[closest_indexes]

    # Take the average value from the K 'y' values that minimizes D
    y_fit = np.array(np.mean(y_closest))

    return xi_closest, y_closest, y_fit, meshgrid_x, y, meshgrid_x_new
```

```python
# Setting K for the KNN method
K = 5

# Number of splits in the input data
N = 100

# x1,x2,...,xn variables
xi = [np.linspace(-5, 5, N) for _ in range(2)] 

# New data point --> x_new = [x1,x2,...,xn]
x_new = np.array([2,-2,]) #  

xi_closest, y_closest, y_fit, meshgrid_x, y, meshgrid_x_new = knn(*xi, x_new=x_new, K=K)
```

### 2D input data visualization

```python
scatter_plot_KNN = go.Scatter3d(
    x=xi_closest[0],
    y=xi_closest[1],
    z=y_closest,
    mode='markers',
    marker=dict(
        size=5,
        color='blue',
        opacity=0.8
    ),
    name='K points'
)

scatter_plot_pred = go.Scatter3d(
    x=meshgrid_x_new[0].flatten(),
    y=meshgrid_x_new[1].flatten(),
    z=y_fit.flatten(),
    mode='markers',
    marker=dict(
        size=5,
        color='red',
        opacity=0.8
    ),
    name='Predicted point'
)

# Create the surface plot
surface_plot = go.Surface(
    x=meshgrid_x[0],
    y=meshgrid_x[1],
    z=y,
    colorscale='plasma',
    opacity=0.3
)

# Create the figure
fig = go.Figure(data=[surface_plot, scatter_plot_KNN, scatter_plot_pred]) # scatter_plot_K

# Update layout
fig.update_layout(
    scene=dict(
        xaxis_title='x1',
        yaxis_title='x2',
        zaxis_title='y',
    )
)

# Show the plot
fig.show()
```
